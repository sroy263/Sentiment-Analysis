```python
def tagging(x):
    clean=[]
    token=[]
    tag=[]
    stops = set(stopwords.words("english"))
    words = [w for w in x if not w in stops]
    for i in words:
        token.append(nltk.word_tokenize(i))
    #token
    for j in token:
        tag.append(nltk.pos_tag(j))
        #tag.append(st.tag(j))
    return tag
```

```python
tag=tagging(x)
tag
```

```python
proper=[]
clean=[]
for i in tag:
    for j in i:
        if(j[1]=='NNP'or j[1]=='RB'or j[1]=='RBR'or j[1]=='RBS'or j[1]=='VB'or j[1]=='JJ'or j[1]=='JJR'or j[1]=='JJS'):
            proper.append(j[0])
    clean.append(proper)
```
